```python
# librerías utilizadas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('datos_chuno_puno.csv')
print(data.head())

```

```python
# identifica valores faltantes
missing_values = data.isnull().sum()
print("Valores faltantes por columna:\n", missing_values)
```

```python
# Verificar si hay valores faltantes
summary_statistics = data.describe()
print("Estadísticas descriptivas:\n", summary_statistics)
```

# Sección nueva

```python
######################   2
# Histograma de la cantidad de papa utilizada
plt.figure(figsize=(10, 6))
sns.histplot(data['Cantidad_Papa_Kg'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Cantidad de Papa Utilizada')
plt.xlabel('Cantidad de Papa (Kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Histograma de la cantidad de chuño
plt.figure(figsize=(10, 6))
sns.histplot(data['Cantidad_Chuno_Kg'], bins=20, kde=True, color='green')
plt.title('Distribución de Cantidad de Chuño')
plt.xlabel('Cantidad de Chuño (Kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de barras para la distribución de la calidad de chuño
plt.figure(figsize=(10, 6))
sns.countplot(x='Calidad', data=data, palette='muted')
plt.title('Distribución de Calidad del Chuño Producido')
plt.xlabel('Calidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de caja para la calidad del chuño
plt.figure(figsize=(8, 6))
sns.boxplot(x='Calidad', y='Cantidad_Chuno_Kg', data=data, palette='Set2')
plt.title('Calidad del Chuño vs Cantidad de Chuño Producida')
plt.xlabel('Calidad del Chuño')
plt.ylabel('Cantidad de Chuño (Kg)')
plt.show()
```

```python
# Gráfico de dispersión entre la temperatura mínima y la cantidad de chuño producido
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperatura_Min', y='Cantidad_Chuno_Kg', hue='Calidad', data=data, palette='viridis')
plt.title('Relación entre Temperatura Mínima y Cantidad de Chuño Producido')
plt.xlabel('Temperatura Mínima (°C)')
plt.ylabel('Cantidad de Chuño (Kg)')
plt.show()
```

```python
# Estadísticas descriptivas de las columnas relevantes
desc_stats = data[['Cantidad_Papa_Kg', 'Cantidad_Chuno_Kg', 'Temperatura_Min', 'Temperatura_Max', 'Humedad']].describe()

desc_stats
```

```python
###################   3
# Identificar las variedades de papa más productivas para el chuño
variedades_productivas = data.groupby('Variedad_Papa')['Cantidad_Chuno_Kg'].sum().reset_index()
variedades_productivas = variedades_productivas.sort_values(by='Cantidad_Chuno_Kg', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Cantidad_Chuno_Kg', y='Variedad_Papa', data=variedades_productivas, palette='viridis')
plt.title('Variedades de Papa más Productivas para el Chuño')
plt.xlabel('Cantidad de Chuño (Kg)')
plt.ylabel('Variedad de Papa')
plt.show()
```

```python
# Comparar la eficacia de los métodos de congelación
eficacia_metodos = data.groupby('Metodo_Congelacion')['Cantidad_Chuno_Kg'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x='Metodo_Congelacion', y='Cantidad_Chuno_Kg', data=eficacia_metodos, palette='Set2')
plt.title('Eficacia de los Métodos de Congelación')
plt.ylabel('Producción Promedio de Chuño (Kg)')
plt.xlabel('Método de Congelación')
plt.show()
```

```python
# Gráfico de caja para visualizar la relación entre el tiempo de congelación y la calidad del chuño
plt.figure(figsize=(10, 6))
sns.boxplot(x='Calidad', y='Tiempo_Congelacion_Dias', data=data, palette='Set2')
plt.title('Relación entre el Tiempo de Congelación y la Calidad del Chuño')
plt.xlabel('Calidad del Chuño')
plt.ylabel('Tiempo de Congelación (días)')
plt.show()
```

```python
############# 4
# Crear una serie temporal de la producción de chuño agrupada por mes y año
data['Mes_Año'] = data['Fecha'].dt.to_period('M')
produccion_mensual = data.groupby('Mes_Año')['Cantidad_Chuno_Kg'].sum()

```

```python
# Graficar la tendencia de la producción de chuño a lo largo del tiempo
plt.figure(figsize=(12, 6))
produccion_mensual.plot()
plt.title('Tendencia de la Producción de Chuño a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Producción de Chuño (Kg)')
plt.grid(True)
plt.show()
```

```python
############  5
# Boxplots para evaluar la relación entre la calidad del chuño y las variables climáticas
plt.figure(figsize=(10, 6))
sns.boxplot(x='Calidad', y='Temperatura_Min', data=data, palette='Set2')
plt.title('Relación entre Temperatura Mínima y Calidad del Chuño')
plt.xlabel('Calidad del Chuño')
plt.ylabel('Temperatura Mínima (°C)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Calidad', y='Temperatura_Max', data=data, palette='Set2')
plt.title('Relación entre Temperatura Máxima y Calidad del Chuño')
plt.xlabel('Calidad del Chuño')
plt.ylabel('Temperatura Máxima (°C)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Calidad', y='Humedad', data=data, palette='Set3')
plt.title('Impacto de la Humedad en la Calidad del Chuño')
plt.xlabel('Calidad del Chuño')
plt.ylabel('Humedad (%)')
plt.show()
```

```python
# Gráfico de dispersión entre humedad y cantidad de chuño producido
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Humedad', y='Cantidad_Chuno_Kg', hue='Calidad', data=data, palette='Set1')
plt.title('Impacto de la Humedad en la Producción de Chuño')
plt.xlabel('Humedad (%)')
plt.ylabel('Cantidad de Chuño Producido (Kg)')
plt.legend(title='Calidad del Chuño')
plt.grid(True)
plt.show()
```
